```python
from scipy.special import gamma
from math import factorial

x = 0.0
mu = 0.0
theta = 20.38

def pdf(x, mu):
    return gamma(x + 1 / theta) / (factorial(x) * gamma(1 / theta)) * (mu / (mu + (1 / theta)))**x * (1 + mu *theta)**(-1 / theta)
```

```python
import altair as alt
import numpy as np
import pandas as pd

mus = np.linspace(0.0, 100, 1000)
d = pd.DataFrame({"mus": mus, "density": [pdf(0, mu) for mu in mus]})
alt.Chart(d).mark_line().encode(x="mus", y="density")
```
